```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Fetch stock data
def fetch_stock_data(ticker, period='1d', interval='1m'):
    stock = yf.Ticker(ticker)
    data = stock.history(period=period, interval=interval)
    return data

# Plot stock price
def plot_stock_price(data, ticker):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Close'], label=f'{ticker} Closing Price')
    plt.title(f'Real-time Stock Price for {ticker}')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # User input for stock ticker
    ticker = input("Enter stock ticker symbol (e.g., AAPL): ").upper()
    
    # Fetch data and plot
    stock_data = fetch_stock_data(ticker)
    plot_stock_price(stock_data, ticker)
```

```python
# Calculate moving averages
def add_moving_averages(data):
    data['MA_10'] = data['Close'].rolling(window=10).mean()
    data['MA_50'] = data['Close'].rolling(window=50).mean()
    return data

# Plot moving averages along with stock price
def plot_stock_with_ma(data, ticker):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Close'], label=f'{ticker} Closing Price')
    plt.plot(data['MA_10'], label='10-period MA')
    plt.plot(data['MA_50'], label='50-period MA')
    plt.title(f'{ticker} Stock Price with Moving Averages')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # Fetch stock data
    stock_data = fetch_stock_data(ticker)
    
    # Add moving averages
    stock_data = add_moving_averages(stock_data)
    
    # Plot with moving averages
    plot_stock_with_ma(stock_data, ticker)
```
